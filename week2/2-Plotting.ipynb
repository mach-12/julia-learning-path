```julia
data = DelimitedFiles.readdlm("ebola_virus_epidemic_major_outbreaks_converted.csv", ',')

epidays = data[:, 1]
allcases = data[:, 2]

data
```

```julia
function convert2int(value)
    # if !isa(value, Int) 
    #     return replace(value, ","=>"")#value
    # end
    # elseif
    #     return value 
    # end

    if isa(value, String) || is(value, SubString{}) == true
        string_value = parse(String, value)
        string_value = replace(value, "\""=>"", ','=>"")

        return parse(Int, string_value)
    
    elseif value =='–'
        return 0
    
    else
        return value
    end
end

# Testing
convert2int_tests = ["10,000", "\"270,000\"", '–', 1099, data[1,2]]

# Iterate through inputs, call convert2int, and print input and output
for test in convert2int_tests
    output = convert2int(test)
    println("Input: $test; Output: $output")
end
```

```julia
for i in 1:size(data, 1)
    for j in 1:size(data, 2)
        println(data[i,j], " ", convert2int(data[i, j]))
    end
end

data
```

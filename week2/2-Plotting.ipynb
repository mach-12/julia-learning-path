# Plotting

## Introduction

- In Julia, We have the Plots Package to deal with plotting operations.

- It is a very flexible package which allows for various diffrent **backends**, which can be triggred without changing the top level **frontend** code.

- There are various Backends which allow for interactive visualization as well!

```julia
using Plots
using DelimitedFiles
```

## Using the `gr` backend

#

```julia
gr()
```

Loading CSV Files

```julia
data = DelimitedFiles.readdlm("ebola_virus_epidemic_major_outbreaks_converted.csv", ',')

epidays = data[:, 1]
allcases = data[:, 2]

data
```

## Plotting curves from coordinate data

```julia
plot(epidays, allcases)
```

```julia
plot(epidays, allcases, linetype= :scatter, marker= :value)
```

```julia
plot(epidays, allcases, 
title = "West African EVD epidemic, total cases",
xlabel = "Days since 22 March 2014",
ylabel = "Total cases to date (three countries)",
marker = (:diamond, 6, "pink"),
line  =  (:path, "gray"),
legend = false,
grid = false
)
```

```julia
savefig("WAfricanEVD_noformatspecified") # default .png
savefig("WAfricanEVD.pdf")
savefig("WAfricanEVD_.png")
```

## Exploring More Parameters

```julia
epidays = data[:, 1]
EVDcasesbycountry = data[:, [4,6,8]]

plot(epidays, EVDcasesbycountry,
legend = :bottomright,
marker = ([:octagon, :star7, :square], 8),
label = ["Guinea" "Liberia" "Sierra Leone"],
title = "West African EVD epidemic cases by country",
xlabel = "Days since 22 March 2014",
ylabel = "Total cases to date",
line = (:scatter)

)
```
